## 1 quantas casas estao disponiveis ?


```python
num_houses_unique = data['id'].nunique()
print('O numero de imooveis disponiveis sao de {}'.format(num_houses_unique))
```

## 2 Qual a casa mais cara? (casa com o maior valor de venda)

```python
# 1. Selecionar a coluna "id", "price", ordenar as casas pela coluna "price" em ordem decrescente 
# 2. escolher o primeiro imovel da lista

house_most_expensive = data[['id', 'price']].sort_values('price', ascending=False).loc[0,'id']
```

```python
print('a casa com maior valor de venda é a de numero: {}'.format(house_most_expensive))
```

## 3 Qual a casa com o maior número de quartos?

```python
# 1. Selecionar a coluna "id", "bedroom"
# 2. Ordenar os imóveis pelo numero de quartos em ordem decrescente
# 3. Selecionar a primeira coluna "id"
df = data[['id', 'bedrooms']].sort_values('bedrooms', ascending=False)
print('casa co maior numero de banheiros: {}'.format(df.iloc[0, 0]))
```

## 4 Quantas casas possuem 2 banheiros?

```python
# 1. Filtrar linhas (imóveis) que possuem 2 banheiros.
# 2. Contar o número de linhas do dataset
df = data.loc[data['bathrooms'] == 2, :]
df
num_houses = len(df)
print('numero de casas com 2 banheiros: {}'.format(num_houses))
```

## 5 Qual o preço médio de todas as casas no conjunto de dados?

```python
# 1. Calcular o preço médio da coluna "price"mean_price = np.round(data['price'].mean(), 2)
mean_price = np.round(data['price'].mean(), 2) # OBS: o numero dois informa a função que sao duas casas decimais no final
mean_price
print('preço medio das casas sao: ${}'.format(mean_price))
```

## 6 Qual o preço médio de casas com 2 banheiros?

```python
# 1. Selecionar imóveis com 2 banheiros.
# 2. Calcular o preço médio da coluna "price" do novo conjunto de dados
media_2_banheiros = np.round(data.loc[data['bathrooms'] == 2,'price'].mean(),2)
print('a media de preço das casas com 2 banheiros é: ${}'.format(media_2_banheiros))
```

## 7 Qual o preço mínimo entre as casas com 3 quartos?

```python
# 1. Selecionar imóveis com 3 bathrooms.
# 2. Calcular o menor preço da coluna "price" do novo conjunto de dados
min_price = np.round(data.loc[data['bedrooms'] == 3, 'price'].min(), 2)
print('casa mais barata com 3 quartos custa: ${}'.format(min_price))
```

##8 Quantas casas possuem mais de 300 metros quadrados na sala de estar?

```python
# 1. Selecionar imóveis com mais de 300 sqft_living.
# 2. Contar o número de imóveis nesse novo conjunto de dados.
houses = data.loc[data['sqft_living'] > 300, 'id'].shape[0]
print('tem {} casas com mais de 300 metros quadrados'.format(houses))
```

##9 Quantas casas tem mais de 2 andares?

```python
# 1. Selecionar imóveis com a coluna 'floors' maior que 2.
# 2. Contar o número de imóveis nesse novo conjunto de dados.
houses = data.loc[data['floors'] > 2, 'id'].shape[0]
print('exietem {} casas com mais de 2 anadares'.format(houses))
```

##10 Quantas casas tem vista para o mar?

```python
# 1. Selecionar imóveis com a coluna 'waterfront' igual a 1.
# 2. Contar o número de imóveis nesse novo conjunto de dados.
houses = data.loc[data['waterfront'] == 1, 'id'].shape[0]
print('existem {} com mais de 2 andares'.format(houses))
```

##filtrando casas para comprar

```python
#filtrando dataframe
condicao1 = data.loc[data['bedrooms'] == 4]
condicao2 = condicao1.loc[data['floors'] == 2.0]
condicao3 = condicao2.loc[data['bathrooms'] == 3]
condicao4 = condicao3.loc[data['sqft_living'] > 2430]
condicao5 = condicao4.loc[data['sqft_lot'] >=6700]
condicao6 = condicao5.loc[data['price'] < 500000]
condicao7 = condicao6.loc[data['price_areaLiv'] < 200]
condicao8 = condicao7.loc[data['yr_built'] > 2000]
condicao9 = condicao8.loc[data['price_sqft_lot'] > 50]
condicao9
```

#*casas escolhidas para comprar*

```python
#5104511600
#5104511840
#1858600042
#5104511630
#3343901401
```

```python
#localizando as casas selecionadas
df_map = flium.Map(olocation=(47.4034, -122.167), zoom_start=10)
for index,row in condicao9.iterrows():
  folium.Marker(location =(row['lat'], row['long']),popup=row['id']).add_to(df_map)
display(df_map)
```
